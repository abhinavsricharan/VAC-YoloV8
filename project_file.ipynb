```python
%pip install opencv-python
import cv2
%pip install ultralytics
%pip install YOLO

# Load YOLOv8 model
model = r"C:\Users\abhin\Downloads\yolov8n.pt"  # Ensure you have the YOLOv8 weights file in the same directory or provide the full path

# Vehicle timings (seconds per vehicle type)
vehicle_timings = {'car': 5, 'truck': 7, 'motorbike': 3, 'bicycle': 3, 'bus': 10}

# Function to process frames and count vehicles
def process_frame_and_count(frame, vehicle_count):
    # Run YOLOv8 inference on the frame
    results = model(frame)
    
    # Iterate through detections
    for result in results:
        boxes = result.boxes  # Bounding boxes
        for box in boxes:
            # Extract bounding box and class information
            x1, y1, x2, y2 = map(int, box.xyxy[0])  # Coordinates
            conf = box.conf[0]  # Confidence
            cls = int(box.cls[0])  # Class index
            label = model.names[cls]  # Class label

            # Filter for specific vehicle classes
            if label in vehicle_timings:
                vehicle_count[label] += 1
                # Draw bounding box and label
                cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 255, 0), 2)
                cv2.putText(frame, f"{label} {conf:.2f}", (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)
    
    return frame, vehicle_count

# Main function to handle input
def main(input_source):
    vehicle_count = {key: 0 for key in vehicle_timings.keys()}  # Initialize vehicle counts

    if input_source.lower().endswith(('.png', '.jpg', '.jpeg')):
        # Process single image
        frame = cv2.imread(input_source)
        if frame is None:
            print("Error: Unable to read the image.")
            return
        
        # Process and display the frame
        frame, vehicle_count = process_frame_and_count(frame, vehicle_count)
        cv2.imshow("Vehicle Detection", frame)
        cv2.waitKey(0)
        cv2.destroyAllWindows()
    
    else:
        # Process video
        cap = cv2.VideoCapture(input_source)
        if not cap.isOpened():
            print("Error: Unable to open the video file.")
            return

        while cap.isOpened():
            ret, frame = cap.read()
            if not ret:
                break
            
            # Process and display the frame
            frame, vehicle_count = process_frame_and_count(frame, vehicle_count)
            cv2.imshow("Vehicle Detection", frame)

            # Break on 'q' key press
            if cv2.waitKey(1) & 0xFF == ord('q'):
                break

        cap.release()
        cv2.destroyAllWindows()
    
    # Calculate total time for each vehicle type
    total_time = {key: count * time for key, (count, time) in zip(vehicle_count.keys(), vehicle_timings.items())}
    
    # Print results
    print("Vehicle Count:", vehicle_count)
    print("Total Time (seconds):", total_time)

# Run the program
if __name__ == "__main__":

    input_path = r"C:\Users\abhin\Downloads\4K Road traffic video for object detection and tracking.mp4 ".strip()
    main(input_path)
```
